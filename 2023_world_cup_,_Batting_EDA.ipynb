**Data Analysis**

we can use some function to check the stats of a match.

we will see ,

* Highest run score in a match
* Highest Strike Rate in a match
* Highest No of 6s in a match
* Highest No of 4s in a match



```python
highest_score_index = df['Runs'].idxmax()
highest_score_batsman_name = df.loc[highest_score_index, 'Batsman_Name']
highest_score = df.loc[highest_score_index, 'Runs']
highest_score_team=df.loc[highest_score_index,'Team_Innings']

print(f"----> The Highest run score in a game was {highest_score} by {highest_score_batsman_name} from {highest_score_team}.")

df['Strike_Rate'] = pd.to_numeric(df['Strike_Rate'], errors='coerce')
highest_strike_rate_index = df['Strike_Rate'].idxmax()
highest_strike_rate_batsman_name = df.loc[highest_score_index, 'Batsman_Name']
highest_strike_rate = df.loc[highest_score_index, 'Strike_Rate']
highest_strike_rate_team=df.loc[highest_score_index,'Team_Innings']

print(f"----> The Highest Strike Rate in a game was {highest_strike_rate} by {highest_score_batsman_name} from {highest_strike_rate_team}.")

most_6_index = df['6s'].idxmax()
most_6_batsman_name = df.loc[highest_score_index, 'Batsman_Name']
most_6 = df.loc[highest_score_index, '6s']
most_6_team=df.loc[highest_score_index,'Team_Innings']

print(f"----> The Highest no of 6s in a game was {most_6} by {most_6_batsman_name} from {most_6_team}.")

most_4_index = df['4s'].idxmax()
most_4_batsman_name = df.loc[highest_score_index, 'Batsman_Name']
most_4 = df.loc[highest_score_index, '4s']
most_4_team=df.loc[highest_score_index,'Team_Innings']

print(f"----> The Highest no of 4s in a game was {most_4} by {most_4_batsman_name} from {most_4_team}.")
```

**As we can see after ruuning the code , Glenn Maxwell has upper hand in each of them . He is the Highest run scorer in a single mactch , most 6s,4s in a match and as well as the strike rate .**

 ***

In above dataframe the table was sorted by each match and each player , Now Create a another dataframe in which the data set is sorted by player name soo we can get a better idea of performance of players in the world cup.

let's sort the data by Batsman name and add the columns Runs, 6s, 4s, etc using group by function.  



```python
df_players=df.groupby('Batsman_Name').agg({'Runs':'sum','4s':'sum','6s':'sum','Balls':'sum','Strike_Rate':'mean','Batting_Position':'mean'})
df_players['Batting_Position'] = df_players['Batting_Position'].astype(int)
df_players_run=df_players.sort_values(by='Runs',ascending=False)
df_players_run.reset_index()
```

So their are 145 batsman in total , this would a large data to make some visualization,so we will take the top 30 batsman based on runs .

* Top run scorer

```python
most_runs=df_players_run.groupby('Batsman_Name')['Runs'].sum().sort_values(ascending=False).reset_index().head(30)
```

```python
plt.figure(figsize=(20,7))
plt.bar(most_runs['Batsman_Name'],most_runs['Runs'],color='orange')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.title('Top Run Scorer in Worldcup 2023')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

**As we can see by the above bar plot the top Run Scorer in this World cup is Virat Kohli followed by Quinton de Kock and Rohit Sharma .**

***

* Most number of 6s

```python
most_6s=df_players_run.groupby('Batsman_Name')['6s'].sum().sort_values(ascending=False).reset_index().head(30)
```

```python
plt.figure(figsize=(20,7))
plt.bar(most_6s['Batsman_Name'],most_6s['6s'],color='green')
plt.xlabel('Players')
plt.ylabel('No of 6s')
plt.title('Top 6s scorer in Worlcup 2023')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

**The most No of 6s is hit by Rohit Sharma followed by David Warner and Quinton de Kock.**

***

* Most number of 4s

```python
most_4s=df_players_run.groupby('Batsman_Name')['4s'].sum().sort_values(ascending=False).reset_index().head(30)
```

```python
plt.figure(figsize=(20,7))
plt.bar(most_4s['Batsman_Name'],most_4s['4s'])
plt.xlabel('Players')
plt.ylabel('No of 4s')
plt.title('Top 6s scorer in Worlcup 2023')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

**The Most no of 4s is hit by Virat kohli followed by Rohit sharma and Quinton de Kock.**

***

* Avg Best Strike Rate

```python
best_strike_rate=df_players_run.groupby('Batsman_Name')['Strike_Rate'].sum().sort_values(ascending=False).reset_index().head(30)
```

```python
plt.figure(figsize=(20,7))
plt.bar(best_strike_rate['Batsman_Name'],best_strike_rate['Strike_Rate'],color='violet')
plt.xlabel('Players')
plt.ylabel('Strike Rate')
plt.title('Best Strike Rate in World Cup 2023')
plt.xticks(rotation=60)
plt.tight_layout()

plt.show()
```

**The avg Best Strike rate is of David willey followed by Aiden Markram and Hardik pandya**

**We have done some analysis based on players, now let's plot some charts based on team performance.**

* Create a data frame in which the stats is sorted on the basis of teams.

```python
df_team=df.groupby('Team_Innings').agg({'Runs':'sum','4s':'sum','6s':'sum','Balls':'sum','Strike_Rate':'mean'})
df_team=df_team.sort_values(by='Runs',ascending=False)
df_team.reset_index()
```

***


* let's Plot a  chart to see which team had scored the most runs.

```python
teams = ["Australia", "South Africa", "India", "New Zealand", "Pakistan", "England", "Afghanistan", "Bangladesh", "Sri Lanka", "Netherlands"]

plt.figure(figsize=(8, 8))
plt.pie(df_team['Runs'], labels=teams, autopct='%1.1f%%')
plt.axis('equal')

plt.title('Runs scored by each team')

plt.show()
```

**As we can see Australia , south africa , India and New zeland had score more than half of the runs from the total runs scored in the torenment.**

```python
teams = ["Australia", "South Africa", "India", "New Zealand", "Pakistan", "England", "Afghanistan", "Bangladesh", "Sri Lanka", "Netherlands"]

plt.plot(teams,df_team['Runs'], marker='o', linestyle='-')
plt.title('Runs scored by each Team')
plt.xlabel('Country')
plt.ylabel('Runs')

plt.xticks(rotation=45)

plt.show()
```

**The top run scorer team is  Australia Followed by south africa and India.**

***

* let's check the 6s and 4s distribution.

```python
teams = ["Australia", "South Africa", "India", "New Zealand", "Pakistan", "England", "Afghanistan", "Bangladesh", "Sri Lanka", "Netherlands"]

barWidth = 0.35
r = np.arange(len(teams))

plt.bar(r,df_team['4s'], color='b', edgecolor='grey', width=barWidth, label='4s')
plt.bar(r,df_team['6s'], bottom=df_team['4s'], color='r', edgecolor='grey', width=barWidth, label='6s')

plt.xlabel('teams', fontweight='bold')
plt.xticks(r,teams, rotation=90)
plt.ylabel('Count')
plt.title('Distribution of 4s and 6s scored by each team')
plt.legend()

plt.show()
```

**In the above chart we can see the of 6s and 4s Scored by each team.**

***

* let's see the Runs made by team based on their Strike rate.

```python
teams = ["Australia", "South Africa", "India", "New Zealand", "Pakistan", "England", "Afghanistan", "Bangladesh", "Sri Lanka", "Netherlands"]
runs = df_team['Runs']
strike_rates = df_team['Strike_Rate']

plt.scatter(runs, strike_rates)
plt.title('Runs vs. Strike Rates for Each Team')
plt.xlabel('Runs')
plt.ylabel('Strike Rate')

for i, team in enumerate(teams):
    plt.text(runs[i], strike_rates[i], team)

plt.show()
```

**By the Above Scatter plot we can see that India has made the runs with Highest Strike Rate follwed by New zeland and Australia.**
